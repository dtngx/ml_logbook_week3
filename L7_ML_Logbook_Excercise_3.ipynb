```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse


## data points
data = [(23, 41), (34, 45), (45, 49), (56, 67), (67, 84), (78, 100)]
x_val = [x[0] for x in data]
y_val = [x[1] for x in data]

## regression line and y-hat calculation
x = np.asarray(x_val)
y = (1.12 * x) + 7.93

## plot the given regression line and data points
fig, ax = plt.subplots(1,1, figsize=(10, 10))
ax.plot(x, y, lw=2)
ax.scatter(x_val, y_val, color="tab:red", marker="x")
ax.grid(True, ls = "--", alpha = 0.5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$y$", fontsize=18)
ax.set_title("Exercise 3.1", fontsize=18)
ax.tick_params(labelsize=14)
ax.legend([r"$y = 1.12 x + 7.93$", "datapoints"], fontsize=14, loc="center left")


## Exercise 3a, calculation of MSE respecting the mean of all datapoints (in this case 6NN)
arraymean = np.full((np.shape(y_val)), np.mean(y_val))
print(f"The MSE respecting the 6NN (or mean) of the data points equals: {mse(y_val, arraymean)}")

## Exercise 3b, calculation of MSE respecting the given regression model
print(f"The MSE respecting the given regression model y = 1.12 * x + 7,93 equals: {mse(y_val, y)}")
```
