```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse


## data points
data = [(23, 41), (34, 45), (45, 49), (56, 67), (67, 84), (78, 100)]
x_val = [x[0] for x in data]
y_val = [x[1] for x in data]

## regression line and y-hat calculation
x = np.asarray(x_val)
y = (1.12 * x) + 7.93

## 
fig, ax = plt.subplots(1,1, figsize=(10, 10))
ax.plot(x, y, lw=1)
ax.scatter(x_val, y_val)
ax.grid(True, ls = "--", alpha = 0.5)

mse(y_val, y)
i = 0
for value in y_val:
  i = value + i
newmean = i / 6
arraymean = [(newmean), (newmean), (newmean), (newmean), (newmean), (newmean)]

mse(y_val, arraymean)
```
